```python
import nltk
import pandas as pd
dataset=pd.read_csv('tweets.csv',error_bad_lines=False)
```

```python
dataset
```

```python
pat=r'''#\w+'''
tokenizer=nltk.regexp_tokenize(str(dataset),pat)
```

```python
tokenizer
```

```python
from nltk.stem import PorterStemmer
ps= PorterStemmer()
tokenizer=list(dict.fromkeys(tokenizer))
```

```python
tokenizer
```

```python
stemmer=[]
for tokens in tokenizer:
    stemmer.append(ps.stem(tokens))
```

```python
stemmer
```

```python
from nltk import ngrams
n=3
sed=ngrams(str(dataset).split(),n)
grams=[]
for s in sed:
    grams.append(s)
```

```python
grams
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
token=RegexpTokenizer(r'[a-zA-Z0-9]+')
cv=CountVectorizer(lowercase=True,ngram_range=(1,1),tokenizer = token.tokenize)
text_counts = cv.fit_transform(tokenizer)
df=pd.DataFrame(text_counts.toarray(),columns=cv.get_feature_names())
df
```
